```python
import numpy as np
# Output: B, C, P, D
def tile_mnist(imgs, shape=(4,4)):
    # B, C, H, W
    width = int(imgs.shape[-2] / shape[1])
    height = int(imgs.shape[-1] / shape[0])
    patches = []
    patch_idx = 0
    for x in range(shape[1]):
        for y in range(shape[0]):
#             patch = imgs[..., 0]
            patch_idx += 1
            patch = imgs[..., y*height:y*height+height, x*width:x*width+width]
            patches.append(patch)
    return np.stack(patches, axis=1)

patched = tile_mnist(x_train[5:100].reshape(-1, 1, 28, 28))
patched.shape
```

```python
patched = tile_mnist(x_train[5:100].reshape(-1, 1, 28, 28), shape=(7,7))
patched.shape
```

```python
grid_shape = (4,4)
patched = tile_mnist(x_train[0:100].reshape(-1, 1, 28, 28), shape=grid_shape)

patch_shapes = patched.shape[-2:]

f, axarr = plt.subplots(grid_shape[0], grid_shape[1])
for x in range(grid_shape[1]):
    for y in range(grid_shape[0]):
        patch_idx = int(x + y*grid_shape[0])
        axarr[x,y].imshow(patched[0, patch_idx, 0])
```
